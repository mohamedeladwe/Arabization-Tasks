```python
import nltk
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix 
```

```python
data = pd.read_csv("processed_data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.Category.unique()
```

# Data Preprocessing

```python
data.drop(["Title"] ,axis =1 ,inplace= True)
data.drop(["Link"] ,axis =1 ,inplace= True)
data.dropna(inplace = True)
```

```python
data
```

```python
#nltk.download('punkt')
```

```python
le = LabelEncoder()
```

```python
data['Category'] =le.fit_transform(data.Category)
```

```python
data
```

# Data Normalization

```python
#data['Content'] = data.Content.map(lambda x: x.lower())
#data['Content'] = data.Content.str.replace('[^\w\s]', '')
```

Tokenization

```python
data['Content'] = data['Content'].apply(nltk.word_tokenize)
```

```python
data['Content'][8396]
```

Stemming

```python
st = nltk.ISRIStemmer()
data['Content'] = data['Content'].apply(lambda x: [st.stem(y) for y in x])
```

```python
data['Content'][8396]
```

Join

```python
data['Content'] = data['Content'].apply(lambda x: ' '.join(x))
```

# Feature Extraction

```python
binary_enc = CountVectorizer(binary = True)
features = binary_enc.fit_transform(data['Content'])
#pd.DataFrame(features.toarray(), columns=tfidf_vect.vocabulary_.keys())
```

```python
x_train,x_test ,y_train,y_test =train_test_split(features, data['Category'] ,test_size =.2 ,shuffle= False)
```

```python
print(x_train.shape)
print(x_test.shape)
```

# Naive Bayes Model

```python
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
predicted = model.predict(x_test)
```

```python
print(np.mean(predicted == y_test))
```

```python
print(confusion_matrix(y_test,predicted))
```
